```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

## Part 1 - XOR
### 1\. Using the XOR dataset below, train (400 epochs) a neural network (NN) using 1, 2, 3, 4, and 5 hidden layers (where each layer has only 2 neurons). For each n layers, store the resulting loss score along with n. Plot the results to find what the optimal number of layers is.

```python
n = 40
xx = np.random.random((n,1))
yy = np.random.random((n,1))
```

```python
X = np.array([np.array([xx,-xx,-xx,xx]),np.array([yy,-yy,yy,-yy])]).reshape(2,4*n).T
y = np.array([np.ones([2*n]),np.zeros([2*n])]).reshape(4*n)
```

```python
plt.scatter(*zip(*X), c=y)
```

### 2-neuron model with 1 hidden layer

```python
model_1 = Sequential()

model_1.add(Dense(2, input_dim=2, activation='tanh')) #hidden layer with hyperbolic tangent as activation
model_1.add(Dense(1, input_dim=2, activation='sigmoid')) #output layer (outputs probability)

sgd = SGD(lr=0.1)
model_1.compile(loss='binary_crossentropy', optimizer='sgd') 

model_1.fit(X, y, batch_size=2, epochs=400) #160/4 = 40 per epoch

print(model_1.predict_proba(X).reshape(4*n))# evaluate the model
```

```python
scores_1 = model_1.evaluate(X, y)
scores_1, model_1.metrics_names
```

```python
plt.scatter(*zip(*X), c=model_1.predict_classes(X))
```

**Observation:** Although the shape of the data is close to our original XOR dataset, the classifications are very off. The loss score isn't in the desired range at all, with a score of 0.331.  The model's poor job at classifying is evident in the plot above.

### 2-neuron model with 2 hidden layers

```python
model_2 = Sequential()

model_2.add(Dense(2, input_dim=2, activation='tanh'))
model_2.add(Dense(2, activation='sigmoid'))
model_2.add(Dense(1, input_dim=2, activation='sigmoid')) #last  output layer must be 1 neuron as it represents the output which is one value, y


sgd = SGD(lr=0.1)
model_2.compile(loss='binary_crossentropy', optimizer='sgd') 

model_2.fit(X, y, batch_size=2, epochs=400) #160/4 = 40 per epoch

print(model_2.predict_proba(X).reshape(4*n))# evaluate the model
```

```python
scores_2 = model_2.evaluate(X, y)
scores_2, model_2.metrics_names
```

```python
plt.scatter(*zip(*X), c=model_2.predict_classes(X))
```

**Observation:** The loss score did not improve, as the score increased to 0.469 from 0.331. The plot shows a worsened model with nearly two halves of the data split (not the 4 corners).

### 2-neuron model with 3 hidden layers

### 4\. Instead of `SGD` try other optimizers and report on the loss score. (https://keras.io/optimizers/)

```python
from keras.optimizers import Adam, RMSprop, Adagrad
```

**Note:** Using the model that didn't perform as well-- 3-neuron, 4-hidden layer, softplus activation (resulted in a loss score of 8.05)-- and seeing how a different optimizer will affect performance.

```python
newmodel4_sp_adam = Sequential()

newmodel4_sp_adam.add(Dense(3, input_dim=2, activation='softplus'))
newmodel4_sp_adam.add(Dense(3, activation='softplus')) 
newmodel4_sp_adam.add(Dense(3, activation='softplus'))
newmodel4_sp_adam.add(Dense(3, activation='softplus'))
newmodel4_sp_adam.add(Dense(1, activation='sigmoid')) 


adam = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
newmodel4_sp_adam.compile(loss='binary_crossentropy', optimizer='Adam') 

newmodel4_sp_adam.fit(X, y, batch_size=2, epochs=400) #160/4 = 40 per epoch

print(newmodel4_sp_adam.predict_proba(X).reshape(4*n))# evaluate the model
```

```python
newscores4_sp_adam = newmodel4_sp_adam.evaluate(X, y)
newscores4_sp_adam, newmodel4_sp_adam.metrics_names
```

### 3. Find a subject/object relationship through the dependency parser in any sentence. 

**Note:** For this portion, let's use the pr_tree against our sentences.  We can find the roots and the dependencies in our processed.text2 sentences.

```python
#first, how many sentences are we looking at in our article's text?

sentences = [s for s in processed_text2.sents]
print(len(sentences)) #length of sentences
```

```python
#print out the 46 sentences

n = 0
for sentence in processed_text2.sents:
    print(n, sentence)
    n+=1
```

```python
#explore the various Parts of Speech and words used in the article

n = 0
for sentence in processed_text2.sents:
    for token in sentence:
        print(n, token, token.pos_, token.lemma_)
        n+=1
```

```python
#let's do a dependency parsing tree for the entire article

def pr_tree(word, level):
    if word.is_punct:
        return
    for child in word.lefts:
        pr_tree(child, level+1)
    print('\t'* level + word.text + ' - ' + word.dep_)
    for child in word.rights:
        pr_tree(child, level+1)
```

```python
#print results of the parsing tree for the article

for sentence in processed_text2.sents:
    pr_tree(sentence.root, 0)
    print('=====================Next Tree=================')
```

**Observation:** This shows the various relationships throughout the article.  The fourth tree, references this sentence: "And bad judgment is what Simpson admitted as a cause for his botched attempt to reclaim intimate family photos and other memorabilia." 

It's interesting this tree pulls out the following subjects: Simpson and judgment.  These are then joined by the following verbs: admitted and the following modifiers: bad, botched.  Towards the end of the sentence it shows some sort of redemtion, where his admittance of "bad judgment" is a cause for his "botched" (modifies attempt) attempt (object) to reclaim (verb/action) his former fame ("photos and other memorabilia" - also other objects in the sentence).

**Note:** Let's take a closer look at this, just by taking out the first sentence and seeing the subject/object relationships.

### 6. Find the most similar noun chunks in the article.

**Note:** Best to use word vectorization to compare similarity among noun_chunks. Let's compare similarity to a select few of the noun chunks, which are added to our comparison vocabulary list.

```python
#first append all the noun chunks into an empty list, in order to then run the similarity scores
n=0
nounlist = []
for noun_chunk in processed_text2.noun_chunks:
    nounlist.append(noun_chunk)
    print(n, noun_chunk)
    n+=1
```

```python
#create a vocabuilary list, the list of words we're comparing our word in the noun chunk list to
#based on the example in the lecture notes e.g., fruit = proc_fruits.vocab['fruit']

athletes = processed_text2.vocab['athletes'] #noun chunk 67
team = processed_text2.vocab['team'] #noun chunk 85
thugs = processed_text2.vocab['thugs'] #noun chunk 92
futureplayers = processed_text2.vocab['Future players'] #noun chunk 109
stardom = processed_text2.vocab['stardom'] #noun chunk 159
```

```python
#now take the noun chunk list and iterate its similarity scores through the list of vocabulary words.  
#save the results into a list (similaritylist)

n=0
similaritylist = []
for word in nounlist:
    similaritylist.append(word)
    similaritylist.append(word.similarity(athletes))
    similaritylist.append(word.similarity(team))
    similaritylist.append(word.similarity(thugs))
    similaritylist.append(word.similarity(futureplayers))
    similaritylist.append(word.similarity(stardom))

    n+=1

similaritylist
```
